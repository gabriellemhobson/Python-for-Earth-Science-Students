```python
import netCDF4 as nc4
import numpy as np
import matplotlib.pyplot as plt
```

```python
sst = nc4.Dataset("20210605120000-C3S-L4_GHRSST-SSTdepth-OSTIA-GLOB_ICDR2.1-v02.0-fv01.0.nc", 'r')
```

```python
print(sst)
```

```python
lon_1d = sst.variables['lon'][:]
lat_1d = sst.variables['lat'][:]
sst_data = sst.variables['analysed_sst'][:]
```

```python
sst.close()
```

```python
lon_1d.shape
```

```python
lat_1d.shape
```

```python
sst_data.shape
```

```python
nlon = len(lon_1d)
nlat = len(lat_1d)
sst = np.reshape(sst_data, (nlat, nlon))
sst.shape
```

```python
fig = plt.figure(frameon=False)
im1 = plt.imshow(sst,origin='lower')
fig.colorbar(im1);
```

```python
fig = plt.figure(frameon=False)
im1 = plt.imshow(sst[2000:3000,2000:3000],origin='lower')
fig.colorbar(im1);
```
